# 1.The Chi-Square Goodness-of-Fit Test


______________

_____________________________________________
- **Null Hypothesis (H0):**  There is no significant association between seeking treatment for a mental health condition and observing negative consequences for coworkers with mental health conditions in the workplace.

- **Alternative Hypothesis (H1):**  There is a significant association between seeking treatment for a mental health condition and observing negative consequences for coworkers with mental health conditions in the workplace.
_________________________________________________

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load the dataset
url = "/population.csv"
df = pd.read_csv(url)

# Select relevant columns for the Chi-Squared Test
columns_for_chi2 = ['family_history', 'treatment', 'no_employees', 'remote_work', 'tech_company',
                    'care_options', 'leave', 'mental_health_consequence', 'phys_health_consequence',
                    'mental_vs_physical', 'obs_consequence']

# Create a contingency table
contingency_table = pd.crosstab(df['treatment'], df['obs_consequence'])

# Perform Chi-Squared Test
chi2, p, _, _ = chi2_contingency(contingency_table)

# Display results
print("Chi-Squared Statistic:", chi2)
print("P-value:", p)

# Interpret the results
alpha = 0.05
print("\nSignificance level:", alpha)
print("Is the p-value less than the significance level?")
if p < alpha:
    print("Yes, there is a significant association between treatment and observed consequences.")
else:
    print("No, there is no significant association between treatment and observed consequences.")
```

# 2. The Chi-square test- test for Normality


_____________

_____________________________________________
- **Null Hypothesis (H0):** There is no significant relationship between the two categorical variables (e.g., 'treatment' and 'mental_health_consequence').

- **Alternative Hypothesis (H1):**  There is a significant relationship between the two categorical variables (e.g., 'treatment' and 'mental_health_consequence').
_________________________________________________

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load the dataset
url = "/population.csv"
data = pd.read_csv(url)

# Select the relevant columns for the Chi-square test
selected_columns = ['treatment', 'mental_health_consequence']
df_chi = data[selected_columns]

# Create a contingency table
contingency_table = pd.crosstab(df_chi['treatment'], df_chi['mental_health_consequence'])

# Perform the Chi-square test
chi2, p, _, _ = chi2_contingency(contingency_table)

# Print the results
print(f"Chi-square value: {chi2}")
print(f"P-value: {p}")

# Check the significance level (usually 0.05)
alpha = 0.05
if p < alpha:
    print("Reject the null hypothesis. There is a significant relationship between 'treatment' and 'mental_health_consequence'.")
else:
    print("Fail to reject the null hypothesis. There is no significant relationship between 'treatment' and 'mental_health_consequence'.")
```

# 3.Chi-square test for independence

__________

_______________________
- **Null Hypothesis (H0):** There is no significant association between seeking treatment for a mental health condition ("treatment") and observing negative consequences for coworkers with mental health conditions ("obs_consequence").

- **Alternative Hypothesis (H1):** There is a significant association between seeking treatment for a mental health condition and observing negative consequences for coworkers with mental health conditions.
___________________________________

```python
import pandas as pd
from scipy.stats import chi2_contingency

# Load the dataset
url = "/population.csv"
data = pd.read_csv(url)

# Select the relevant columns for the Chi-square test
selected_columns = ['treatment', 'obs_consequence']
contingency_table = pd.crosstab(data[selected_columns[0]], data[selected_columns[1]])

# Perform the Chi-square test
chi2_stat, p, _, expected = chi2_contingency(contingency_table)

# Display the results
print("Chi-square test for independence:")
print("\nStep 1: Null Hypothesis (H0): There is no significant association between seeking treatment and observing negative consequences for coworkers.")
print("     Alternative Hypothesis (H1): There is a significant association between seeking treatment and observing negative consequences.")
print("\nStep 2: Chi-square test results")
print(f"     Chi-square value: {chi2_stat}")
print(f"     P-value: {p}")

# Display expected values
print("\nStep 3.1: Expected Values")
print(expected)

# Calculate the test statistic
test_statistic = ((contingency_table - expected) ** 2 / expected).sum().sum()
print(f"\nStep 3.2: Test Statistic (Chi-square): {test_statistic}")

# Step 4: Make the decision
alpha = 0.05
df = (contingency_table.shape[0] - 1) * (contingency_table.shape[1] - 1)  # degrees of freedom
critical_value = chi2.ppf(1 - alpha, df)

print("\nStep 4: Decision")
print(f"     Critical Value: {critical_value}")

if chi2_stat > critical_value:
    print("     Reject the null hypothesis. There is a significant association.")
else:
    print("     Fail to reject the null hypothesis. There is no significant association.")

# Step 5: Summarize the results
print("\nStep 5: Summary")
print(f"     Degrees of Freedom: {df}")
print(f"     Decision: {'Reject' if chi2_stat > critical_value else 'Fail to reject'} the null hypothesis.")
```

# 4. Chi-square test of Homogenity

_________________

_____________________________
- **Null Hypothesis (H0):** The distribution of seeking treatment is the same across different types of companies.

- **Alternative Hypothesis (H1):** The distribution of seeking treatment is different across different types of companies.
__________________________________

```python
import pandas as pd
from scipy.stats import chi2_contingency, chi2

# Load the dataset
url = "/population.csv"
data = pd.read_csv(url)

# Select the relevant columns for the Chi-square test of homogeneity
selected_columns = ['treatment', 'tech_company']
contingency_table = pd.crosstab(data[selected_columns[0]], data[selected_columns[1]])

# Perform the Chi-square test of homogeneity
chi2_stat, p, _, expected = chi2_contingency(contingency_table)

# Display the results
print("Chi-square test of homogeneity:")
print("\nStep 1: Null Hypothesis (H0): The distribution of seeking treatment is the same across different types of companies.")
print("     Alternative Hypothesis (H1): The distribution of seeking treatment is different across different types of companies.")
print("\nStep 2: Chi-square test results")
print(f"     Chi-square value: {chi2_stat}")
print(f"     P-value: {p}")

# Display expected values
print("\nStep 3.1: Expected Values")
print(expected)

# Calculate the test statistic
test_statistic = ((contingency_table - expected) ** 2 / expected).sum().sum()
print(f"\nStep 3.2: Test Statistic (Chi-square): {test_statistic}")

# Step 4: Make the decision
alpha = 0.05
df = (contingency_table.shape[0] - 1) * (contingency_table.shape[1] - 1)  # degrees of freedom
critical_value = chi2.ppf(1 - alpha, df)

print("\nStep 4: Decision")
print(f"     Critical Value: {critical_value}")

if chi2_stat > critical_value:
    print("     Reject the null hypothesis. The distribution of seeking treatment is different across different types of companies.")
else:
    print("     Fail to reject the null hypothesis. The distribution of seeking treatment is the same across different types of companies.")

# Step 5: Summarize the results
print("\nStep 5: Summary")
print(f"     Degrees of Freedom: {df}")
print(f"     Decision: {'Reject' if chi2_stat > critical_value else 'Fail to reject'} the null hypothesis.")
```

# 6. Yates correction for continuity in a 2x2 contingency table

_________________

_____________________________
- **Null Hypothesis (H0):** There is no association between seeking treatment for a mental health condition and observing negative consequences for coworkers with mental health conditions.
- **Alternative Hypothesis (H1):**  There is an association between seeking treatment for a mental health condition and observing negative consequences for coworkers with mental health conditions.
__________________________________

```python
import pandas as pd
from scipy.stats import chi2

# Load the dataset
url = "/population.csv"
data = pd.read_csv(url)

# Select the relevant columns for the Chi-square test for a 2x2 contingency table
selected_columns = ['treatment', 'obs_consequence']
contingency_table = pd.crosstab(data[selected_columns[0]], data[selected_columns[1]])

# Step 1: State the hypotheses and identify the claim
# Null Hypothesis (H0): There is no association between seeking treatment and observing negative consequences.
# Alternative Hypothesis (H1): There is an association between seeking treatment and observing negative consequences.
print("Step 1: State the hypotheses and identify the claim")
print("     Null Hypothesis (H0): There is no association between seeking treatment and observing negative consequences.")
print("     Alternative Hypothesis (H1): There is an association between seeking treatment and observing negative consequences.")

# Step 2: Find the critical value in the right tail
alpha = 0.05
df = 1  # degrees of freedom for a 2x2 contingency table
critical_value = chi2.ppf(1 - alpha, df)
print("\nStep 2: Find the critical value in the right tail")
print(f"     Critical Value: {critical_value}")

# Step 3: Compute the test value using Yates correction for continuity
N = contingency_table.sum().sum()
test_value = abs(N * ((contingency_table.iloc[0, 0] * contingency_table.iloc[1, 1]) - (contingency_table.iloc[0, 1] * contingency_table.iloc[1, 0]) - N / 2) ** 2)
test_value /= ((contingency_table.iloc[0, :].sum() + N) * (contingency_table.iloc[1, :].sum() + N) * (contingency_table.iloc[:, 0].sum() + N) * (contingency_table.iloc[:, 1].sum() + N))
print("\nStep 3: Compute the test value")
print(f"     Test Value: {test_value}")

# Step 4: Make the decision
print("\nStep 4: Make the decision")
if test_value > critical_value:
    print("     Reject the null hypothesis. There is an association between seeking treatment and observing negative consequences.")
else:
    print("     Fail to reject the null hypothesis. There is no association between seeking treatment and observing negative consequences.")

# Step 5: Summarize the results
print("\nStep 5: Summarize the results")
print(f"     Degrees of Freedom: {df}")
print(f"     Decision: {'Reject' if test_value > critical_value else 'Fail to reject'} the null hypothesis.")
```
