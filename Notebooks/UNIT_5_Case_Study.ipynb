# 1.Sample size calculation

______________

_____________
# For single mean n= z**2(alpha**2)/d**2

______________

_____________________________________________
- **Null Hypothesis (H0):**  There is no significant difference in the willingness of employees in the tech industry to seek mental health treatment, regardless of their demographics or workplace characteristics.

- **Alternative Hypothesis (H1):**  There is a significant difference in the willingness of employees in the tech industry to seek mental health treatment based on their demographics or workplace characteristics.
_________________________________________________

```python
import math

# Given values
alpha = 0.05  # desired level of significance
z_score = 1.96  # Z-score for the desired level of significance
margin_of_error = 0.05  # desired margin of error

# Sample size calculation for a single mean
sample_size = (z_score**2 * alpha**2) / margin_of_error**2

# Round up to the nearest whole number, as the sample size must be an integer
sample_size = math.ceil(sample_size)

# Output steps
print("Step 1: Define Parameters")
print(f"   - Desired level of significance (alpha): {alpha}")
print(f"   - Z-score for level of significance (z_score): {z_score}")
print(f"   - Desired margin of error: {margin_of_error}")

print("\nStep 2: Calculate Sample Size")
print(f"   Sample size = ({z_score}^2 * {alpha}^2) / {margin_of_error}^2")

# Display intermediate values
intermediate_value = (z_score**2 * alpha**2)
print(f"   Intermediate value = {intermediate_value}")

# Display the final sample size
print(f"   Sample size = {sample_size}")

# Round up to the nearest whole number, as the sample size must be an integer
sample_size = math.ceil(sample_size)

# Output the final result
print(f"\nStep 3: Round Up")
print(f"   Rounded up sample size: {sample_size}")

```

# For single proportion n=z**2(p*(1-p))/d**2==z**2*p*q/E**2


______________


____________________________________________________

*   **Null Hypothesis:** There is no significant association between demographic or workplace factors and the likelihood of seeking treatment for mental health conditions among employees in the tech industry.
*   **Alternative Hypothesis:**  There is a significant association between one or more demographic or workplace factors and the likelihood of seeking treatment for mental health conditions among employees in the tech industry.
____________________________________________________

# 2. Sample size for two samples
For two means N=n1+n2=4 sd**2(z(1-alpha/2)+z(1-beta))**2/(d=mu1-mu2)**2

_____________

_____________________________________________
- **Null Hypothesis (H0):** There is no significant difference in the mean [dependent variable] between [group 1] and [group 2] in the tech industry.

- **Alternative Hypothesis (H1):**  There is a significant difference in the mean [dependent variable] between [group 1] and [group 2] in the tech industry.
_________________________________________________

```python
import math

# Given values
alpha = 0.05  # desired level of significance
beta = 0.2  # desired power (1 - beta)
z_alpha_over_2 = 1.96  # Z-score for the desired level of significance
z_beta = 0.84  # Z-score for the desired power

# Placeholder for effect size - replace with actual effect size based on your study
effect_size = 0.5  # Adjust this based on your study

# Sample size calculation for two independent samples
sample_size = (z_alpha_over_2 + z_beta)**2 * (2 * effect_size**2) / (alpha**2)

# Round up to the nearest whole number, as the sample size must be an integer
sample_size = math.ceil(sample_size)

# Output steps
print("Step 1: Define Parameters")
print(f"   - Desired level of significance (alpha): {alpha}")
print(f"   - Desired power (1 - beta): {beta}")
print(f"   - Z-score for level of significance (z_alpha_over_2): {z_alpha_over_2}")
print(f"   - Z-score for power (z_beta): {z_beta}")
print(f"   - Effect size: {effect_size}")

print("\nStep 2: Calculate Sample Size")
print(f"   Sample size = ({z_alpha_over_2} + {z_beta})^2 * (2 * {effect_size}^2) / {alpha}^2")

# Display intermediate values
intermediate_value = (z_alpha_over_2 + z_beta)**2 * (2 * effect_size**2)
print(f"   Intermediate value = {intermediate_value}")

# Display the final sample size
print(f"   Sample size = {sample_size}")

# Round up to the nearest whole number, as the sample size must be an integer
sample_size = math.ceil(sample_size)

# Output the final result
print(f"\nStep 3: Round Up")
print(f"   Rounded up sample size: {sample_size}")
```

# 3.For two proportions
N=n1+n2=
4(z(1-alpha/2)+z(1-beta))**2)[(P1+P2)/2(1-(P1+P2)/2))]/(d=mu1-mu2)**2
__________

_______________________
- **Null Hypothesis (H0):** There is no significant difference in the proportions of [attribute of interest] between [group 1] and [group 2] in the tech industry.

- **Alternative Hypothesis (H1):** There is a significant difference in the proportions of [attribute of interest] between [group 1] and [group 2] in the tech industry.
___________________________________

```python
import math

# Given values
alpha = 0.05  # desired level of significance
beta = 0.2  # desired power (1 - beta)
z_alpha_over_2 = 1.96  # Z-score for the desired level of significance
z_beta = 0.84  # Z-score for the desired power

# Placeholder for estimated proportions - replace with actual estimates based on your study
p1_estimate = 0.3  # Adjust this based on your study
p2_estimate = 0.5  # Adjust this based on your study

# Placeholder for effect size - replace with actual effect size based on your study
effect_size = 0.1  # Adjust this based on your study

# Sample size calculation for two proportions
N = ((z_alpha_over_2 + z_beta)**2 * ((p1_estimate + p2_estimate) / 2) * (1 - (p1_estimate + p2_estimate) / 2)) / effect_size**2

# Round up to the nearest whole number, as the sample size must be an integer
N = math.ceil(N)

# Output steps
print("Step 1: Define Parameters")
print(f"   - Desired level of significance (alpha): {alpha}")
print(f"   - Desired power (1 - beta): {beta}")
print(f"   - Z-score for level of significance (z_alpha_over_2): {z_alpha_over_2}")
print(f"   - Z-score for power (z_beta): {z_beta}")

print("\nStep 2: Estimate Proportions")
print(f"   Estimated proportion 1 (p1): {p1_estimate}")
print(f"   Estimated proportion 2 (p2): {p2_estimate}")

print("\nStep 3: Calculate Sample Size")
print(f"   N = (({z_alpha_over_2} + {z_beta})^2 * (({p1_estimate} + {p2_estimate}) / 2) * (1 - ({p1_estimate} + {p2_estimate}) / 2)) / {effect_size}^2")

# Display intermediate values
intermediate_value = ((z_alpha_over_2 + z_beta)**2 * ((p1_estimate + p2_estimate) / 2) * (1 - (p1_estimate + p2_estimate) / 2))
print(f"   Intermediate value = {intermediate_value}")

# Display the final sample size
print(f"   N = {N}")

# Round up to the nearest whole number, as the sample size must be an integer
N = math.ceil(N)

# Output the final result
print(f"\nStep 4: Round Up")
print(f"   Rounded up sample size (N): {N}")
```
