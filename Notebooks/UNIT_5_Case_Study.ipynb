```python
# Function to get Z-value for power
def get_Z_power(power):
    if power == 0.8:
        return 0.842
    elif power == 0.9:
        return 1.282
    elif power == 0.95:
        return 1.645
    elif power == 0.99:
        return 2.326
    else:
        raise ValueError("Invalid power. Use 0.8, 0.9, 0.95, or 0.99.")

def get_Z_TAB(significance_level, side):
    # Define the Z-table values
    z_table = {
        0.01: {'two_sided': 2.576, 'one_sided': 2.326},
        0.05: {'two_sided': 1.960, 'one_sided': 1.645},
        0.10: {'two_sided': 1.645, 'one_sided': 1.282}
    }

    # Check if the significance level is in the table
    if significance_level in z_table:
        # Check if the side is valid
        if side in ['two_sided', 'one_sided']:
            return z_table[significance_level][side]
        else:
            raise ValueError("Invalid side. Use 'two_sided' or 'one_sided'.")
    else:
        raise ValueError("Invalid significance level. Use 0.01, 0.05, or 0.10.")
```

# 1.Sample size calculation

_____________
# For single mean n= z**2(alpha**2)/d**2

______________

_____________________________________________
- **Null Hypothesis (H0):**  There is no significant difference in the willingness of employees in the tech industry to seek mental health treatment, regardless of their demographics or workplace characteristics.

- **Alternative Hypothesis (H1):**  There is a significant difference in the willingness of employees in the tech industry to seek mental health treatment based on their demographics or workplace characteristics.
_________________________________________________

```python
import pandas as pd
import math

# Function to estimate sample size for one-sample problem
def estimate_sample_size_one_sample(df, variable, margin_of_error, confidence_level=0.05):
    # Extract the data for the specified variable
    data = df[variable]

    # Calculate the estimated population standard deviation
    population_std_dev = data.std()

    # Get the Z-value for the specified confidence level
    z_value = get_Z_TAB(confidence_level, 'two_sided')

    # Formula for sample size estimation for one-sample problem
    sample_size = math.ceil((z_value**2 * population_std_dev**2) / range_of_error**2)

    return sample_size

# Load your dataset
df = pd.read_csv('data.csv')

# Example usage:
variable_of_interest = 'Age'  # Replace with the variable you want to estimate
range_of_error = 2.0  # Specify the desired margin of error

# Estimate sample size for one-sample problem
sample_size = estimate_sample_size_one_sample(df, variable_of_interest, range_of_error)

print(f"The estimated sample size needed for the one-sample estimation of '{variable_of_interest}' is: {sample_size}")
```

# For single proportion n=z**2(p*(1-p))/d**2==z**2*p*q/E**2

In a study focusing on mental health within the tech industry, there is an interest in estimating the proportion of individuals with a specific mental health consequence. In a previous study conducted at another tech company, a proportion of 30% of employees was found to have this mental health consequence.

```python
import pandas as pd
import math

# Function to estimate sample size for one-sample proportion
def estimate_sample_size_one_proportion(p, margin_of_error, alpha):
    # Get the Z-value for the specified confidence level
    z_value = get_Z_TAB(alpha, 'two_sided')

    # Formula for sample size estimation for one-sample proportion
    sample_size = math.ceil((z_value**2 * p * (1 - p)) / margin_of_error**2)

    return sample_size

# Load your dataset
df = pd.read_csv('data.csv')

# Example usage:
estimated_proportion = 0.4  # Replace with the estimated proportion from the previous study
margin_of_error = 0.09  # Specify the desired margin of error
alpha = 0.05
# Estimate sample size for one-sample proportion
sample_size = estimate_sample_size_one_proportion(estimated_proportion, margin_of_error,alpha)

print(f"The estimated minimal sample size needed for the one-sample proportion estimation is: {sample_size}")
```

# 2. Sample size for two samples
_____________

```python
import pandas as pd
import math

# Function to estimate sample sizes for two-sample comparison
def estimate_sample_sizes_two_sample(sigma, alpha, power, d):
    # Get Z-values for the specified significance and power levels
    z_alpha_over_2 = get_Z_TAB(alpha, 'two_sided')
    z_power = get_Z_power(power)

    # Formula for sample size estimation for two-sample comparison
    N = math.ceil((4 * sigma**2 * (z_alpha_over_2 + z_power)**2) / d**2)

    # Calculate individual sample sizes (n1 and n2)
    n1 = n2 = math.ceil(N / 2)

    return n1, n2

# Load your dataset
df = pd.read_csv('data.csv')

# Example usage:
sigma = 30.0  # Specify the estimated standard deviation
alpha = 0.05  # Significance level
power = 0.90  # Desired power
d = 20.0  # Desired difference between means

# Estimate sample sizes for two-sample comparison
n1, n2 = estimate_sample_sizes_two_sample(sigma, alpha, power, d)

print(f"The estimated sample size for group 1 (n1) is: {n1}")
print(f"The estimated sample size for group 2 (n2) is: {n2}")
```

# 4. For two proportions
N=n1+n2=
4(z(1-alpha/2)+z(1-beta))**2)[(P1+P2)/2(1-(P1+P2)/2))]/(d=mu1-mu2)**2

Null hypothesis: Mu1=mu2
_________________

_____________________________
- **Null Hypothesis (H0):** The proportions of [attribute of interest] are equal between [group 1] and [group 2] in the tech industry.

- **Alternative Hypothesis (H1):** The proportions of [attribute of interest] are not equal between [group 1] and [group 2] in the tech industry.
__________________________________

```python
import pandas as pd
import math

# Function to estimate sample sizes for two-sample proportion test
def estimate_sample_sizes_two_sample_proportion(p1, p2, alpha, power, d):
    # Get Z-values for the specified significance and power levels
    z_alpha_over_2 = get_Z_TAB(alpha,"two_sided")
    z_power = get_Z_power(power)

    # Formula for sample size estimation for two-sample proportion test
    N = math.ceil(4 * (z_alpha_over_2 + z_power)**2 * ((p1 + p2) / 2 * (1 - (p1 + p2) / 2)) / d**2)

    # Calculate individual sample sizes (n1 and n2)
    n1 = n2 = math.ceil(N / 2)

    return n1, n2

# Specify the expected proportions for each group (p1 and p2)
p1_expected = 0.75  # Adjust based on your hypothesis
p2_expected = 0.50  # Adjust based on your hypothesis

# Specify the significance level (alpha), desired power, and expected difference (d)
alpha_value = 0.05
power_value = 0.9
expected_difference = p1_expected - p2_expected

# Estimate sample sizes for two-sample proportion test
n1_estimate, n2_estimate = estimate_sample_sizes_two_sample_proportion(p1_expected, p2_expected, alpha_value, power_value, expected_difference)

print(f"The estimated sample size for group 1 (n1) is: {n1_estimate}")
print(f"The estimated sample size for group 2 (n2) is: {n2_estimate}")
```
