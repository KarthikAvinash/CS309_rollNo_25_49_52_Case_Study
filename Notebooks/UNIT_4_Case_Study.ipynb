# 1.The Chi-Square Goodness-of-Fit Test


______________

```python
def get_chi_square_tab(alpha, dof):
    # Tabulated chi-square values
    chi_square_table = {
        0.10: {1: 2.706, 2: 4.605, 3: 6.251, 4: 7.779, 5: 9.236, 6: 10.645, 7: 12.017, 8: 13.362, 9: 14.684, 10: 15.987},
        0.05: {1: 3.841, 2: 5.991, 3: 7.815, 4: 9.488, 5: 11.071, 6: 12.592, 7: 14.067, 8: 15.507, 9: 16.919, 10: 18.307},
        0.01: {1: 6.635, 2: 9.210, 3: 11.345, 4: 13.277, 5: 15.086, 6: 16.812, 7: 18.475, 8: 20.090, 9: 21.666, 10: 23.209}
    }

    # Check if alpha and dof are in the table
    if alpha in chi_square_table and dof in chi_square_table[alpha]:
        return chi_square_table[alpha][dof]
    else:
        print("Invalid alpha or degrees of freedom.")
        return None
```

```python
import pandas as pd
df = pd.read_csv('data.csv')

# Check the unique values in the 'mental_vs_physical' column
unique_values = df['mental_vs_physical'].unique()

# Initialize a dictionary to store observed frequencies
observed_frequency_dict = {}

# Count the occurrences of each category
for value in unique_values:
    observed_frequency_dict[value] = df[df['mental_vs_physical'] == value].shape[0]

# Display the observed frequencies
print("Observed Frequencies:")
n = 0
for key, value in observed_frequency_dict.items():
    n = n + 1
    print(f"{key}: {value}")

dof = n - 1
print("\n\nDegrees of Freedom: ", dof)

alpha = 0.05

# Calculate the chi-square critical value
chi_tab = get_chi_square_tab(alpha=alpha, dof=dof)
print("Chi-square Critical Value: ", chi_tab)

# Create a DataFrame for the observed frequencies
observed_df = pd.DataFrame(list(observed_frequency_dict.items()), columns=['mental_vs_physical', 'observed_frequency'])

# Define the expected frequency under the null hypothesis
expected_frequency = observed_df['observed_frequency'].sum() / len(observed_df)

# Calculate the expected frequencies for each category
expected_frequencies = [expected_frequency] * len(observed_df)
observed_df['expected_frequency'] = expected_frequencies

# Calculate the chi-squared statistic
chi_stat = sum((observed_df['observed_frequency'] - expected_frequency) ** 2 / expected_frequency)

# Display the expected frequencies
print("\nExpected Frequencies:")
for key, value in observed_frequency_dict.items():
    print(f"{key}: {expected_frequency}")

# Display the chi-squared statistic
print(f"\nChi-squared statistic: {chi_stat}")

# Compare the chi-squared statistic with the critical value
if chi_stat > chi_tab:
    print("Reject the null hypothesis. There is enough evidence to suggest a significant difference in the distribution.")
else:
    print("Fail to reject the null hypothesis. There is not enough evidence to suggest a significant difference in the distribution.")
```

# 2.Chi-square test for Normality

__________

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Load your dataset
df = pd.read_csv('data.csv')
df = df[df["treatment"]=="Yes"]
# Extract the 'Age' column
age_data = df['Age']

# Define the number of bins for the histogram
num_bins = 10

# Calculate observed frequencies and bin edges
observed_frequencies, bin_edges = np.histogram(age_data, bins=num_bins)
observed_frequencies = observed_frequencies.astype(float)

# Calculate expected frequencies (assuming normal distribution)
mean_age = age_data.mean()
std_dev_age = age_data.std()
expected_frequencies = np.diff(stats.norm.cdf(bin_edges, mean_age, std_dev_age)) * len(age_data)

# Perform the chi-squared test for normality
chi_squared_stat = np.sum((observed_frequencies - expected_frequencies)**2 / expected_frequencies)

# Degrees of freedom
dof = num_bins - 1

# Significance level
alpha = 0.05

# Critical value from chi-squared distribution table
chi_tab = get_chi_square_tab(alpha, dof)

# Display the results
print("Chi-squared statistic for normality:", chi_squared_stat)
print("Degrees of Freedom:", dof)
print("Critical value from chi-squared distribution table:", chi_tab)

# Create a DataFrame for observed and expected frequencies
freq_table = pd.DataFrame({
    'Bin Edges': bin_edges[:-1],
    'Observed Frequencies': observed_frequencies,
    'Expected Frequencies (assuming normal distribution)': expected_frequencies
})

# Display the table
print("\nObserved and Expected Frequencies:")
print(freq_table)

# Compare the chi-squared statistic with the critical value
if chi_squared_stat > chi_tab:
    print("\nReject the null hypothesis. There is evidence that the 'Age' column is not normally distributed.")
else:
    print("\nFail to reject the null hypothesis. There is no significant evidence that the 'Age' column is not normally distributed.")

# Plot the histogram
# plt.hist(age_data, bins=num_bins, edgecolor='black', alpha=0.7, label='Observed Frequencies')
# plt.plot(bin_edges[:-1], expected_frequencies, linestyle='dashed', color='red', label='Expected Frequencies (Normal Distribution)')
# plt.xlabel('Age')
# plt.ylabel('Frequency')
# plt.title('Histogram and Expected Frequencies for Age')
# plt.legend()
# plt.show()
```

# 3.Chi-square test for independence

__________

whether there is a difference in opinion between employees who work remotely and those who work on-site.

```python
# Load your dataset
df = pd.read_csv('data.csv')

# Display the unique values in the 'remote_work' column
print("Unique values in 'remote_work' column:", df['remote_work'].unique())

# Create a contingency table
contingency_table = pd.crosstab(df['remote_work'], df['mental_vs_physical'])

# Display the contingency table
print("\nContingency Table:")
print(contingency_table)

# Calculate observed frequencies
observed_frequencies = contingency_table.values

# Calculate the chi-squared statistic
chi_stat = sum((observed_frequencies[i, j] - contingency_table.sum(axis=1)[i] * contingency_table.sum(axis=0)[j] / contingency_table.sum().sum())**2 / (contingency_table.sum(axis=1)[i] * contingency_table.sum(axis=0)[j] / contingency_table.sum().sum()) for i in range(contingency_table.shape[0]) for j in range(contingency_table.shape[1]))

# Degrees of freedom calculation
dof = (contingency_table.shape[0] - 1) * (contingency_table.shape[1] - 1)

# Significance level
alpha = 0.05

# Get chi-square critical value
chi_tab = get_chi_square_tab(alpha, dof)

# Display the results
print(f"\nChi-squared statistic: {chi_stat}")
print(f"Degrees of Freedom: {dof}")
print(f"Chi-squared critical value: {chi_tab}")

# Compare the chi-squared statistic with the critical value
if chi_stat > chi_tab:
    print("Reject the null hypothesis. There is evidence of a significant association between remote work and mental health perception.")
else:
    print("Fail to reject the null hypothesis. There is no significant association between remote work and mental health perception.")
```
